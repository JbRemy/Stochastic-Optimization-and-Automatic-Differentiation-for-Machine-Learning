# <center>Stochastic Optimization and Automatic Differentiation for Machine Learning</center>

```python
from SVGR import SGD, SVRG, SVRG2, CM, AM

from memory_profiler import memory_usage
import numpy as np
import matplotlib.pyplot as plt
```

First of all let's test each algorithm on a toy data set. With high dimensionality. 

```python
X = np.random.normal(size=(10000, 1000))
p = 0.01
Theta = np.random.choice([0, 1], size=(1000,), p=[1-p, p])
y = np.dot(X, Theta) + np.random.normal(scale=10, size=(10000,))
```

```python
sgd = SGD()
mem_usage_sgd = memory_usage((sgd.fit, ([y, X],), {'K' : 100, 'T' : 2000,
                                                   'learning_rate' : 0.001,
                                                   'save_at_every_step' : False,
                                                   'Theta_0' : np.random.choice([0.0, 1.0], size=(1000,), p=[1-p, p])}),
                             interval=.01)
```

```python
svrg = SVRG()
mem_usage_svrg = memory_usage((svrg.fit, ([y, X],), {'K' : 5, 'T' : 500,
                                                     'learning_rate' : 0.001,
                                                     'save_at_every_step' : True}), interval=.01)
```

```python
svrg2 = SVRG2()
mem_usage_svrg2 = memory_usage((svrg2.fit, ([y, X],), {'K' : 5, 'T' : 500,
                                                      'learning_rate' : 0.001,
                                                      'save_at_every_step' : True}), interval=.01)
```

```python
cm = CM()
mem_usage_cm = memory_usage((cm.fit, ([y, X],), {'K' : 5, 'T' : 500,
                                                 'learning_rate' : 0.001,
                                                 'save_at_every_step' : True}), interval=.01)
```

```python
am = AM()
mem_usage_am = memory_usage((am.fit, ([y, X],), {'K' : 5, 'T' : 500,
                                                 'learning_rate' : 0.001,
                                                 'save_at_every_step' : True}), interval=.01)
```

### Performance comparison :

```python
#Memory consuption

fig = plt.figure(figsize=(10,10))

plt.plot(mem_usage_sgd)
plt.plot(np.array(mem_usage_svrg) - mem_usage_sgd[-1])
plt.plot(np.array(mem_usage_svrg2) - mem_usage_svrg[-1])
plt.plot(np.array(mem_usage_am) - mem_usage_svrg2[-1])
plt.plot(np.array(mem_usage_cm) - mem_usage_am[-1])

plt.legend(['sgd', 'svrg', 'svrg2', 'am', 'cm'])

plt.show()
```

```python
#Memory consuption

fig = plt.figure(figsize=(10,10))

plt.plot(sgd.time_memory, sgd.loss_memory)
plt.plot(svrg.time_memory, svrg.loss_memory)
plt.plot(svrg2.time_memory, svrg2.loss_memory)
plt.plot(am.time_memory, am.loss_memory)
plt.plot(cm.time_memory, cm.loss_memory)

plt.xscale('log')

plt.legend(['sgd', 'svrg', 'svrg2', 'am', 'cm'])

plt.show()
```

```python
def moving_average(a, n=50) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

```python
#Memory consuption

fig = plt.figure(figsize=(10,10))

#plt.plot(sgd.loss_memory)
plt.plot(moving_average(svrg.loss_memory[0:750]))
plt.plot(moving_average(svrg2.loss_memory[0:750]))
plt.plot(moving_average(am.loss_memory[0:750]))
plt.plot(moving_average(cm.loss_memory[0:750]))

#plt.legend(['sgd', 'svrg', 'svrg2', 'am', 'cm'])
plt.legend(['svrg', 'svrg2', 'am', 'cm'])

plt.show()
```

```python
#Memory consuption

fig = plt.figure(figsize=(10,10))

#plt.plot(sgd.loss_memory)
plt.plot(svrg.time_memory[0:701], moving_average(svrg.loss_memory[0:750]))
plt.plot(svrg2.time_memory[0:701], moving_average(svrg2.loss_memory[0:750]))
plt.plot(am.time_memory[0:701], moving_average(am.loss_memory[0:750]))
plt.plot(cm.time_memory[0:701], moving_average(cm.loss_memory[0:750]))

#plt.legend(['sgd', 'svrg', 'svrg2', 'am', 'cm'])
plt.legend(['svrg', 'svrg2', 'am', 'cm'])

plt.show()
```
